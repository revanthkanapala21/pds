```python
import pandas as pd
```

```python
data_frame = pd.read_csv('../clean_data/StudentsPerformance.csv')
print(data_frame.to_string())
```

```python
data_frame["Average"]=(data_frame["math score"]+data_frame["reading score"]+data_frame["writing score"])/3
print(data_frame["Average"])
```

```python
def Grade(Average):
    if (Average >= 95):return 'O'
    if (Average >= 81):return 'A'
    if (Average >= 71):return 'B'
    if (Average >= 61):return 'C'
    if (Average >= 51):return 'D'
    if (Average >= 41):return 'E'
    else: return 'F'
    
data_frame["grade"] = data_frame.apply(lambda x : Grade(x["Average"]),axis=1)
```

```python
import seaborn as sns
import numpy as np
```

```python
ax = sns.countplot(y="grade", hue="lunch", data=data_frame, order=["O", "A", "B", "C", "D", "E", "F"])
fig = ax.get_figure()
fig.savefig("../results/Analysis_Grade_Lunch.png")
```

```python
ax = sns.countplot(y="grade", hue="race/ethnicity", data=data_frame, order=["O", "A", "B", "C", "D", "E", "F"])
fig = ax.get_figure()
fig.savefig("../results/Analysis_Grade_Race.png")
```
